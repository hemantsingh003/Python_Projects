### Drop Column

```python
#use drop for remove the unwanted columns or rows from the data set
df.drop("index", axis=1, errors='ignore', inplace=True)
df.drop("Job Description", axis=1, errors='ignore', inplace=True)
```

```python
#use df to show data set according to past changes
df
```

### Set default Index from 0 to 1

```python
#use index to change the starting no. of the index
df.index = range(1, len(df) + 1)
df
```

### Remove Garbage Items from Company Name

```python
#use lambda function to create dynamic function for cleaning 
df["Company Name"] = df["Company Name"].apply(lambda x: x.split('\n')[0])
df[["Company Name"]]
```

### Remove text like 'Glassdoor est.', '$', 'K', '/yr'

```python
#use copy for transfer same data to different column
df["SALARY"] = df["Salary Estimate"].copy()
```

```python
#use replace to remove garbage character
df["SALARY"] = (df["SALARY"]
                .str.replace(r"\(Glassdoor est.\)","",regex = True)
                .str.replace(r"[$]","",regex = True)
                .str.replace(r"[K]","", regex = True)
                .str.replace(r"/yr","", regex = True)
                .str.strip())
```

```python
df["Salary Estimate"] = df["SALARY"]
```

```python
df= df.drop("SALARY",axis = 1)
df[["Salary Estimate"]]
```

### Split Salary to Min & Max Salary

```python
#use split for break salary on the bases of "-"
df[["Min Salary (in Thousands)","Max Salary (in Thousands)"]] = df["Salary Estimate"].str.split("-",expand = True)
df.drop("Salary Estimate", axis=1, errors='ignore', inplace=True)
df[["Min Salary (in Thousands)","Max Salary (in Thousands)"]]
```

### Convert to Numeric Data Type

```python
#use to_numeric for convert text data type to numerical
df["Min Salary (in Thousands)"] = pd.to_numeric(df["Min Salary (in Thousands)"], errors="coerce")
df["Max Salary (in Thousands)"] = pd.to_numeric(df["Max Salary (in Thousands)"], errors="coerce")
df[["Min Salary (in Thousands)","Max Salary (in Thousands)"]].info()
```

### Calculate Avg. Salary

```python
#use mean for calculate the average salary, here are min and max salary 
df["Avg Salary"] = df[["Min Salary (in Thousands)","Max Salary (in Thousands)"]].mean(axis = 1)
df[["Avg Salary"]]
```

### Delete Duplicate Data

```python
#use shape for check dimension
df.shape
```

```python
#use drop duplicate for remove data which is appear more than one time
df = df.drop_duplicates()
df.index = range(1, len(df) + 1)
df.shape
```

### Standardize Job Titles: Title Case

```python
#use title for provide standard format
df["Job Title"] = df["Job Title"].str.title()
df[["Job Title"]]
```

### Handling Placeholder Values in 'Competitors' Column

```python
#use loop for automate work
processed_competitors = []

for comp in df["Competitors"]:
    if comp == "-1":
        processed_competitors.append("No Competitor")
    else:
        processed_competitors.append(comp)

df["Competitors"] = processed_competitors
df[["Competitors"]]
```

### Which job titles offer the highest average salary?

```python
#use group by for make category of same things
#use sort values for sorting the data like ascending or descending
avg_salary_by_job = df.groupby('Job Title')["Avg Salary"].mean().reset_index()
top_jobs = avg_salary_by_job.sort_values(by="Avg Salary", ascending=False)

print(top_jobs.head(10))
```

### Handle False Value by Replace "-1" to "NaN" 

```python
#replace both type of false like "-1" & 1 
df.replace("-1","NaN",inplace = True)
df["Founded"]=df["Founded"].replace(-1,np.nan)
```
